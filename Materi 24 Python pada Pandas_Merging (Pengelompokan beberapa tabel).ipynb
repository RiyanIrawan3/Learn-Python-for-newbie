```python
df = pd.DataFrame({'keys':['k0','k1','k2','k3','k4'],
                         'One':[1,2,3,4,5],
                         'two':[6,7,8,9,10]})
```

```python
df1 = pd.DataFrame({'keys':['k0','k1','k2','k3','k4'],
                         'three':[1,2,3,4,5],
                         'four':[6,7,8,9,10]})
```

```python
df
```

```python
df1
```

## Merge

```python
New = pd.merge(df,df1, how='inner', on='keys')
New
```

```python
df3 = pd.DataFrame({'keys':['k0','k1','k2','k3','k4'],
                    'keys1':['k0','k1','k0','k3','k1'],
                         'A':[1,2,3,4,5],
                         'B':[6,7,8,9,10]})
```

```python
df4 = pd.DataFrame({'keys':['k1','k1','k0','k3','k4'],
                    'keys1':['k0','k1','k0','k0','k1'],
                         'C':[1,2,3,4,5],
                         'D':[6,7,8,9,10]})
```

```python
df3
```

```python
df4
```

```python
New1 = pd.merge(df3,df4, how='left', on=['keys','keys1'])
New1
```
