```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(2,22,0.8)
y = x**2
```

```python
fig = plt.figure(figsize=(4,5), dpi=100) #dpi itu cara memperbesar, kalo figsize berisi(panjang, tinggi)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
#version 2
```

```python
fig, axes = plt.subplots(2,1, figsize=(3,3), dpi=100)

axes[0].plot(x,y)
axes[0].set_title('Grafik 1 Utama')

axes[1].plot(x,y)
axes[1].set_title('Grafik 2 Utama')

plt.tight_layout()
plt.show()
```

# Menambah kan sebuah lagenda
#### yaitu keterangan pada sebuah grafik

```python
fig = plt.figure(facecolor='green', dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Grafik  1', lw=1, ls='-', marker='o', markersize=4, mfc='green', mew='1', mec='blue')
ax.plot(x,x**4, label='Grafik 2', lw=2, ls='--', marker='*', markersize=4, mfc='green', mew='1', mec='red')
ax.plot(x,x**3, label='Grafik 3', lw=3.6 , ls='-.', marker='+', markersize=4, mfc='green', mew='1', mec='yellow')
ax.grid(True) #untuk menambah grid

ax.legend(loc=(2)) #kita pun bisa merubah ket pada tabel dengan loc
ax.set_xlim([0,11.6])
ax.set_ylim([0,2500])
ax.set_title('Grafik Bullish', color='white')

plt.show()
```
