```python
tabel1 = pd.DataFrame(np.random.rand(5,5),[0,1,2,3,4],['A','B','C','D','E'])
```

```python
tabel1
```

```python
tabel1.plot.bar() #mirip seperti histogram
```

```python
tabel1.plot.bar(stacked=True) #melihat pembagian pada setiap histogram
```

```python
tabel1['B'].plot.bar() #kita pun bisa melihat isi dari histogram B
```

```python
tabel1.plot.scatter(x='A', y='B', c='B', cmap='ocean') 
# c= itu berisakan tabel mana yang mau kita pilih, dan cmap itu warna dari grafiknya. semakin hijau warna makin rendah nilainya
```

```python
tabel1.plot.scatter(x='A', y='B', s=tabel1.D**100, c='C', cmap='ocean')
#itu cara memperbesar titiknya
```

```python
tabel1.plot.box()
# A bawah itu nilai terkecil dan A yang atas nilai terbesar. berisikan mean, median
```

```python
tabel1.plot.line()
```

```python
tabel1.plot.line(y='B', lw=15, figsize=(7,4))
# jadi Y disini itu variabel apa yang kita inginkan. lw disini tebal garisnya. untuk figsize (panjang,lebar)
```

```python
tabel1.plot.hexbin(x='A', y='B', gridsize=40) 
#semakin kecil angkanya semakin besar warnanya. disini kita bisa menganalisis lewat warna
```
