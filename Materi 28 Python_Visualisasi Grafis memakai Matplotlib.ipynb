```python
import numpy as np
x = [12,24,36,49,59]
y = np.array(x)**2
```

```python
plt.plot(x,y);
```

```python
plt.plot(x,y)
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.title('Grafik X dan Y')
plt.show()
```

```python
plt.subplot(121)
plt.plot(x,y,color='red')
plt.xlabel('Sumbu X1')
plt.ylabel('Sumbu Y1')

plt.subplot(122)
plt.plot(x,y,color='blue')
plt.xlabel('Sumbu X2')
plt.ylabel('Sumbu Y2')
plt.tight_layout() #untuk memisahkan antara dua diagram di bawah
plt.show()
```
