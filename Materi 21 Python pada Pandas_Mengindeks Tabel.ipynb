# iloc
## .iloc[baris, kolom]

```python
items.head()
```

```python
items.iloc[0]
```

```python
items.iloc[0:10] #untuk pemelihan baris
```

```python
items.iloc[:,4] #kalo kolomnya brantakan series
```

```python
items.tail(10)
```

```python
items.iloc[-5:-1,-4:] #-4 itu kolom, sedangkan -1 itu yang terqakhir tidak masuk
```

```python
items.iloc[[0,3,7]] #ini cara untuk mengambil data pada kolom
```

```python
items.iloc[[2,3,5,6],[2,3,4]] #jadi ini ada  baris dan kolom 2,3,5,6 ini baris sedangkan 2,3,4 adalah kolom
```

# loc
## loc['nama','nama kolom']

#### untuk melakukan slicing tapi tidak mengetahui indeksnya bagaimana

```python
items = pd.read_csv('kapal_titanic.csv', index_col='embarked')
```

```python
items #diatas cara mengganti nomor kolom jadi string
```

```python
#kita bisa melakukan indexing
items.loc['S']
```

```python
items.loc['S','sex']
```

```python
items.loc['S',['age','fare']]
```

```python
items.loc[['S','Q'],['age','fare']] #kita bisa baris dan kolom pada loc jika tidak mengetahui indkexing
```
