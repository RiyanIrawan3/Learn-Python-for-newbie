```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
```

```python
Data = pd.read_csv('kapal_titanic.csv')
Data.head()
```

```python
Data.describe(include='O')
```

# Barplot

```python
# Barplot (Histogram)
sns.barplot(x='sex', y='fare', data=Data);
```

```python
 sns.barplot(x='fare', y='sex', data=Data);
```

```python
 sns.barplot(x='sex', y='fare', data=Data, estimator=sum);
# buat menambah kan angka pada diagram
```

```python
 sns.barplot(x='sex', y='fare', data=Data, estimator=np.std);
```

# Countplot

```python
# Countplot
sns.countplot(x='sex', data=Data)
```

```python
Data # lihat di atas datanya sekitar 800an
```

```python
sns.countplot(x='deck', hue='sex', data=Data);
```

```python
# grafik di atas tidak urut, maka kita akan mengurutnya
sns.countplot(x='deck', hue='sex', order=['A','B','C','D','E','F','G'], data=Data);
```

# Boxplot

```python
##### boxplot (box & whiskers plot) dengan plot ini kita bisa melihat q1= 25% Data, q2= 50% Data Median, q3=75% Data, q4=100% Data
sns.boxplot(x='sex', y='fare', data=Data);

#### garis yang di luar diagram adalah namanya outlier
#### outlier atas
#### q3 + 1,5 *(q3-q1)
#### oulier bawah
#### q1-1,5*(q3-Q1)
#### garis paling bawah diagram adalah q1-1,5
#### q3 -q1 interquartile Range (IR)
```

```python
sns.boxplot(x='sex', y='fare', hue='deck', hue_order=['A','B','C','D','E','F','G'], data=Data);
```

# Violin Plot

```python
sns.violinplot(x='sex' ,y='fare' , data=Data);
# plot ini menunjukan distribusi
```

```python
sns.violinplot(x='sex', y='fare', hue='deck', hue_order=['A','B','C','D','E','F','G'], data=Data);
# di violin plot kita juga bisa melihat q1, q2, q3 dan batas outlier seperti di boxplot. dan milihat persebaran distribusinya
```

```python
sns.violinplot(x='deck', y='fare', hue='sex', order= ['A','B','C','D','E','F','G'], split=True, data=Data);
# true ini bisa di pakai kalo dikolom tidak lebih dari dua.
```

# Strip Plot

```python
sns.stripplot(x='deck', y='fare', hue='sex', order= ['A','B','C','D','E','F','G'], data=Data);
# pada dasarnya ini scatter plot
# kita mengurtkan bedasarkan melalui hue
```

```python
# kalo kurang jelas dan ingin terlihat disebar maka dengan jitter
sns.stripplot(x='deck', y='fare', hue='sex', order= ['A','B','C','D','E','F','G'], jitter=True, data=Data);
```

# Swarmplot

```python
sns.swarmplot(x='deck', y='fare', hue='sex', order= ['A','B','C','D','E','F','G'], data=Data);
# Pada dasar nya ini scatter plot, tapi mirip violin plot
```

```python
sns.violinplot(x='deck', y='fare', order= ['A','B','C','D','E','F','G'], data=Data);
sns.swarmplot(x='deck', y='fare', order= ['A','B','C','D','E','F','G'], color='blue', data=Data);
```
