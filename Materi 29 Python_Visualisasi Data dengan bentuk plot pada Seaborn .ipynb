```python
import scipy.stats as stat
```

```python
# versi 2 yaitu desstribusi Z(normal)
from scipy.stats import zscore
```

```python
import numpy as np
nilai_Z = zscore(dataku.fare)
```

```python
nilai_Z
```

```python
nilai_Z_abs = np.abs(nilai_Z)
```

```python
nilai_Z_abs #ini angka abssolut ini angka yang ga ada min
```

```python
filter_data = (nilai_Z_abs<3)
```

```python
dataku_baru = dataku[filter_data]
```

```python
dataku_baru #baris tinggal 871
```

```python
sns.jointplot(x='age', y='fare', stat_func=stat.pearsonr, kind='reg', data=dataku_baru);
```

```python
sns.jointplot(x='age', y='fare', stat_func=stat.pearsonr, kind='hex', data=data_baru);
```

```python
sns.jointplot(x='age', y='fare', stat_func=stat.pearsonr, kind='kde', data=data_baru);
```
