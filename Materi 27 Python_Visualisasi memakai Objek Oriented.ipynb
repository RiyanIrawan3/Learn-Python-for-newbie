```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot #cara fungsional dan konvensional
```

```python
x = np.arange(1,25,0.7)
y = x**2
```

```python
x
```

```python
y
```

```python
# Versi matplotlib pada OOP
# Version 1
fig = plt.figure() #sama dengan plt. plot
axes = fig.add_axes([0.2, 0.2, 0.9, 0.9]) #axes object oriented
axes.plot(x,y)
axes.set_xlabel('Sumbu X')
axes.set_ylabel('Sumbu Y')
axes.set_title('Tabel Grafik')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.2, 0.2, 0.9, 0.9])
axes2 = fig.add_axes([0.3, 0.5, 0.4, 0.4])

axes1.plot(x,y, color='blue')
axes1.set_xlabel('Sumbu x Besar')
axes1.set_ylabel('Sumbu y Kecil')
axes.set_title('Grafik Utama')

axes2.plot(y,x, color='red')
axes2.set_xlabel('Sumbu X kecil')
axes2.set_ylabel('Sumbu Y Kecil')
plt.show()
```

```python
# Version 2
# plt.figure()
# add_axes
# kita di version 2 ini akan menggabungkan kedua nya
```

```python
# Version 2
fig, axes = plt.subplots(2,1)#ini ada 2 kolom dan 2 baris

# Plot for the first axes
axes[0].plot(x,y, color='blue')
axes[0].set_xlabel('Sumbu X')
axes[0].set_ylabel('Sumbu Y')
axes[0].set_title('Grafik Utama')

# PLot for the ssecond axes
axes[1].plot(y,x, color='red')
axes[1].set_xlabel('Sumbu X2')
axes[1].set_ylabel('Sumbu Y2')
axes[1].set_title('Grafik Kedua')

fig.tight_layout()
plt.show()
```
